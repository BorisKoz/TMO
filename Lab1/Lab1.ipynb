## визуальный анализ

Давайте теперь построим пару графиков для нескольких параметров.

```python
## пример - флаваноиды и яблочная кислота
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='flavanoids', y='malic_acid', data=data)
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='flavanoids', y='malic_acid', data=data, hue = 'target')
```

Видно, что виды с разными показателями целевого параметра группируются вместе. млжно предположить корелляцию данных параметров.

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data['flavanoids'])
```

```python
sns.jointplot(x='flavanoids', y='malic_acid', data=data)
#видно, что значения флаваноидов распределены более равномерно, а значения яблочной кислоты сконцентрированы в районе [1,2]
```

```python
sns.jointplot(x='flavanoids', y='malic_acid', data=data,  hue = 'target')
```

```python
#построим распределение плотности этих параметров по target:
sns.violinplot(x='target', y='flavanoids', data=data)
```

```python
sns.violinplot(x='target', y='malic_acid', data=data)
```

Можно увидеть, что данные фигуры были также обрисованы в команде sns.jointplot(x='flavanoids', y='malic_acid', data=data,  hue = 'target'). Там они так же обозначают распределение плотности данных для каждой категории target.

```python
sns.pairplot(data, hue='target')
```

Подобный анализ можно провести для каждого из этих графиков. Чем более визуально разделены группы target на графике, тем лучше кореляция.

## кореляционная матрица

```python
data.corr()
```

```python
plt.figure(figsize = (20, 20))
sns.heatmap(data.corr(), annot=True, fmt='.3f')
```

При таком количестве параметров heatmap-не лучший инструмент, но пока еще приемлемый. Можно увидеть сильные кореляции между параметрами target и flavanoids, total_phelons, od280. все эти параметры оказывают большое влияние на target, вместе с тем будучи связаными между собой. При желании можно удалить od280 и total_phenols. Наименьшее влияние на целевой признак оказывает параметр ash, так что его тоже можно удалить. Модель можно использовать для МО, тк между большей частью параметров существуют слабые кореляции (~0.2) и при этом все параметры кроме ash имеют кореляцию в районе 0.5 с target.
